## Import Libraries and Dependencies

```python
#Imports
import ccxt
import pandas as pd
import hvplot.pandas
import os
import time
from Historic_Crypto import HistoricalData
from dotenv import load_dotenv
from bokeh.models.formatters import DatetimeTickFormatter
```

```python
# twilio_client = Client(twilio_account_sid, twilio_auth_token)
exchange_id = 'coinbasepro'
exchange_class = getattr(ccxt, exchange_id)
exchange = exchange_class({
    'timeout': 30000,
    'enableRateLimit': True
})
```

```python
def retrieve_ohlcv(product):
    return exchange.fetch_ohlcv (product, '1m')

def get_current_price(product):
    price = retrieve_ohlcv(product)[-1][4]
    return price

def retrieve_historical_data(product,start_date):
    return HistoricalData(product, 86400, start_date).retrieve_data()
```

```python
# Use the `concat` function to combine the DataFrames by matching indexes (or in this case `time`)
combined_df = pd.concat([bitcoin, litecoin, ethereum, dash, stellar], axis="columns", join="inner")
combined_df.head() 
```

```python
# Drop the extra columns from the DataFrame
combined_df.drop(columns=["open", "high", "low", "volume"], inplace=True)
combined_df
```

```python
# Rename the columns by modifying the columns DataFrame attribute
combined_df.columns = ["bitcoin", "litecoin", "ethereum", "dash", "stellar"]
combined_df
```

# Do you need a Diversified Crypto Portfolio to Spread Risk and does Bitcoin really lift the Altcoin market?


## 1. Are Altcoins more correlated than Bitcoin?

```python
# Plot Altcoins data using hvplot.area 
formatter = DatetimeTickFormatter(months='%b %Y')
combined_df.hvplot.area(y=["litecoin", "ethereum", "dash", "stellar"], xformatter=formatter, title="Altcoins correlation", width=1200, height=600)
```

## 2. Bitcoin and Altcoins correlation?

```python
# Plot Bitcoin and Altcoins data using hvplot.line
combined_df.hvplot.line(x="time", y=["bitcoin", "litecoin", "ethereum", "dash", "stellar"], xformatter=formatter, title="Bitcoin and Altcoins correlation",  width=1200, height=600)
```

```python
#Use the 'pct_change' function to calculate daily returns for each Crypto
daily_returns = combined_df.pct_change()
daily_returns.hvplot.line(groupby=['index.year', 'index.month'], widget_type='scrubber', widget_location='bottom')
```

```python
#combined_df.hvplot.heatmap
correlation = combined_df.corr()
correlation.hvplot.heatmap( 
    x='columns',
    y='index', 
    title='Crypto correlation', 
    cmap=["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce", "#ddb7b1", "#cc7878", "#933b41", "#550b1d"],
    xaxis='top', 
    rot=70,
    width=600, height=600).opts(
    toolbar=None, 
    fontsize={'title': 10, 'xticks': 5, 'yticks': 5}
)
```
